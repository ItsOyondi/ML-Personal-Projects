```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/jobs.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_degree = LabelEncoder()
le_job = LabelEncoder()
le_company = LabelEncoder()
le_location = LabelEncoder()
```

```python
#preprocessing data
df['degree_n'] = le_degree.fit_transform(df['Degree'])
df['company_n'] = le_degree.fit_transform(df['Company'])
df['job_n'] = le_degree.fit_transform(df['Job Title'])
df['location_n'] = le_degree.fit_transform(df['Location'])

df.head()
```

```python
X = df.drop(columns=['Degree','Company', 'Job Title', 'Location', 'Salary'])
y = df['Salary']
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()

model.fit(X, y)
```

```python
model.score(X,y)
```
