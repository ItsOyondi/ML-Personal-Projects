```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]

# Create the plot
plt.plot(x, y)

# Add a title
plt.title('Basic Line Plot')

# Add labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

```python
plt.plot(x, y, color='green', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=12)

plt.title('Customized Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)  # Adds grid lines
plt.show()
```

```python
y2 = [14, 13, 12, 11, 10]

plt.plot(x, y, label='Line 1')
plt.plot(x, y2, label='Line 2')

plt.title('Multiple Lines')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()
plt.show()
```

```python
# Multiple subplots
# Define the figure and axes
fig, axs = plt.subplots(3, 2)

# Plot on different axes
axs[0, 0].plot(x, y)
axs[0, 1].plot(x, y2, 'r')
axs[1, 0].plot(y, x, 'g')
axs[1, 1].plot(y2, x, 'b')
axs[2, 0].plot(y2, x)

# Add titles to subplots
axs[0, 0].set_title('Plot 1')
axs[0, 1].set_title('Plot 2')
axs[1, 0].set_title('Plot 3')
axs[1, 1].set_title('Plot 4')
axs[2, 0].set_title('Plot 4 duplicated')

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
#With annotation
plt.plot(x, y)

plt.title('Plot with Annotations')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Add annotation
plt.annotate('Highest Point', xy=(5, 14), xytext=(3, 12),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
# 3D visualization
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data
z = [1, 2, 3, 4, 5]
x = [10, 11, 12, 13, 14]
y = [15, 14, 13, 12, 11]

# Plot data
ax.plot(x, y, z)

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
# Bar charts
categories = ['A', 'B', 'C']
values = [5, 7, 3]

plt.bar(categories, values)

plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')

plt.show()
```

```python
# Historgrams
data = [1, 1, 2, 2, 2, 3, 3, 4, 5, 5, 5, 5, 6, 7, 8]

plt.hist(data, bins=5)

plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.show()
```

```python
# Scatter plots
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]

plt.scatter(x, y)

plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, gaussian_kde

# Generate random data following a normal distribution
data = np.random.normal(0, 1, 1000)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the KDE (smooth curve)
kde = gaussian_kde(data)
x = np.linspace(min(data), max(data), 1000)
plt.plot(x, kde(x), 'k', linewidth=2)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [6, 5, 4, 3, 2]
}
df = pd.DataFrame(data)

# Set up the subplot grid
num_columns = len(df.columns)
rows = (num_columns + 1) // 2  # Determine the number of rows needed (2 columns per row)
fig, axes = plt.subplots(rows, 2, figsize=(10, 8))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plotting each column in a different subplot using a for loop
for i, col in enumerate(df.columns):
    df[col].plot(ax=axes[i], title=f'Column {col}')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
